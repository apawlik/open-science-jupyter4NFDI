# Open Science with Jupyter4NFDI

Welcome to the **Open Science with Jupyter4NFDI** tutorial.

This repository contains interactive training material introducing key concepts of Open Science and the FAIR principles using Jupyter notebooks. The notebooks are designed to run directly in a web browser using Binder or the Jupyter4NFDI infrastructure, allowing learners to explore concepts, execute code, and complete exercises without installing software locally.

## Learning objectives

- Explain the principles of Open Science.
- Describe the FAIR principles.
- Identify Open Science practices and repositories.
- Explore interactive Jupyter notebooks.
